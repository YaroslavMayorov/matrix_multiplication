# Matrix Multiplication Benchmark Analysis

This notebook analyzes the performance metrics of different matrix multiplication algorithms: **Naive**, **Strassen**, **Winograd**, and **BLAS**.

Data is loaded from `benchmark_results.csv`.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = [10, 6]

# Load data
df = pd.read_csv('benchmark_results.csv')
df.head()
```

## Graph 1: Multiplications vs Matrix Size
**Configuration**: Integer Data, Random Matrices.
**Algorithms**: Naive, Strassen, Winograd (BLAS excluded).

```python
# Filter for Int Random and exclude BLAS
subset = df[(df['DataType Type'] == 'Int Random') & (df['Algorithm'] != 'BLAS')]

plt.figure()
sns.lineplot(data=subset, x='Size', y='Multiplications', hue='Algorithm', marker='o')
plt.title('Multiplications vs Matrix Size (Int Random)')
plt.ylabel('Number of Multiplications')
plt.xlabel('Matrix Size (N)')
plt.xscale('log', base=2)
plt.yscale('log')
plt.show()

# Display Table
pivot_mults = subset.pivot(index='Size', columns='Algorithm', values='Multiplications')
display(pivot_mults)
```

## Graph 2: Additions vs Matrix Size
**Configuration**: Integer Data, Random Matrices.
**Algorithms**: Naive, Strassen, Winograd (BLAS excluded).

```python
# Filter for Int Random and exclude BLAS
subset = df[(df['DataType Type'] == 'Int Random') & (df['Algorithm'] != 'BLAS')]

plt.figure()
sns.lineplot(data=subset, x='Size', y='Additions', hue='Algorithm', marker='o')
plt.title('Additions vs Matrix Size (Int Random)')
plt.ylabel('Number of Additions')
plt.xlabel('Matrix Size (N)')
plt.xscale('log', base=2)
plt.yscale('log')
plt.show()

# Display Table
pivot_adds = subset.pivot(index='Size', columns='Algorithm', values='Additions')
display(pivot_adds)
```

## Graph 3 Series: Runtime Analysis
Comparing Runtime vs Size for all algorithms (including BLAS) across all Data Types and Matrix Types.

```python
# Get unique combinations
combinations = df['DataType Type'].unique()

for combo in combinations:
    subset = df[df['DataType Type'] == combo]
    
    plt.figure()
    sns.lineplot(data=subset, x='Size', y='Time(ms)', hue='Algorithm', marker='o')
    plt.title(f'Runtime vs Size ({combo})')
    plt.ylabel('Time (ms)')
    plt.xlabel('Matrix Size (N)')
    plt.xscale('log', base=2)
    plt.yscale('log')
    plt.grid(True, which="both", ls="--")
    plt.show()
    
    # Display Table
    pivot_time = subset.pivot(index='Size', columns='Algorithm', values='Time(ms)')
    display(pivot_time)
    print("-" * 80)
```

## Graph 4 Series: Memory vs Matrix Size
Comparing Memory usage vs Size for Naive, Strassen, and Winograd (BLAS excluded) across all Data Types and Matrix Types.

```python
# Get unique combinations
combinations = df['DataType Type'].unique()

for combo in combinations:
    subset = df[(df['DataType Type'] == combo) & (df['Algorithm'] != 'BLAS')]
    
    plt.figure()
    sns.lineplot(data=subset, x='Size', y='Memory(KB)', hue='Algorithm', marker='o')
    plt.title(f'Memory vs Size ({combo})')
    plt.ylabel('Memory (KB)')
    plt.xlabel('Matrix Size (N)')
    plt.xscale('log', base=2)
    plt.yscale('log')
    plt.grid(True, which="both", ls="--")
    plt.show()
    
    # Display Table
    pivot_memory = subset.pivot(index='Size', columns='Algorithm', values='Memory(KB)')
    display(pivot_memory)
    print("-" * 80)
```

## Graph 5: Error vs Matrix Size (Complex Matrices)
Comparing numerical error vs Size for Strassen and Winograd algorithms on Complex matrices across all matrix types.

```python
# Filter for Complex data type and algorithms with non-zero error
complex_data = df[df['DataType Type'].str.contains('Complex')]
error_algos = complex_data[complex_data['Algorithm'].isin(['Strassen', 'Winograd'])]

# Get unique matrix types for Complex
matrix_types = error_algos['DataType Type'].unique()

for mtype in matrix_types:
    if mtype == 'Complex Identity':
        continue
    subset = error_algos[error_algos['DataType Type'] == mtype]
    
    plt.figure()
    sns.lineplot(data=subset, x='Size', y='Error', hue='Algorithm', marker='o')
    plt.title(f'Numerical Error vs Size ({mtype})')
    plt.ylabel('Error (Sum of Absolute Differences)')
    plt.xlabel('Matrix Size (N)')
    plt.xscale('log', base=2)
    plt.yscale('log')
    plt.grid(True, which="both", ls="--")
    plt.show()
    
    # Display Table
    pivot_error = subset.pivot(index='Size', columns='Algorithm', values='Error')
    display(pivot_error)
    print("-" * 80)
```
